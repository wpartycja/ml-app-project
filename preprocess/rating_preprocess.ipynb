```python
import pandas as pd
```

```python
movies_df = pd.read_csv('./preprocessed_data/movies_preprocessed.csv')
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_ids = movies_df['id']
movies_ids
```

```python
ratings_df = pd.read_csv('./data/ratings_small.csv')
```

```python
ratings_df.drop(columns=['timestamp'], inplace=True)
```

```python
ratings_df[ratings_df.duplicated()]
```

```python
ratings_df.dropna(inplace=True)
```

```python
ratings_df = ratings_df.loc[ratings_df['movieId'].isin(movies_ids)]
```

```python
def change_rating_scale(rating):
    return rating * 2

ratings_df['rating'] = ratings_df['rating'].apply(lambda x: change_rating_scale(x))
ratings_df['rating'].value_counts()
```

```python
ratings_df.head()
```

```python
ratings_df.to_csv('./preprocessed_data/ratings_small_preprocessed.csv')
```
